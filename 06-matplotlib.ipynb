```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"]#设置字体
plt.rcParams["figure.figsize"]=(6.4,4.8)#设置大小，默认为英寸
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
```

```python
x=[1,2,3,4]
y=[1,4,2,3]
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel('变量a')
ax.set_ylabel('entryb')
plt.show()
```
